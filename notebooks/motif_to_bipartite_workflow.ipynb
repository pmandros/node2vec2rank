```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
#do the data have index and header??
index_col = None
header = None

motif_prior_filename = "cancer_colon_motif.txt"
motif_prior_pd = pd.read_csv('priors/'+motif_prior_filename,  index_col=index_col, header=header, sep="\t")
```

```python
set_of_nodes = set(motif_prior_pd.iloc[:,1]).union(set(motif_prior_pd.iloc[:,2]))
```

```python
motif_prior_pd
```

```python
edges = nx.read_weighted_edgelist('priors/'+motif_prior_filename,delimiter="\t")
```

```python
my_graph = nx.Graph()
my_graph.add_nodes_from(set_of_nodes)
my_graph.add_edges_from(edges.edges())


```

```python
data_np = nx.to_pandas_adjacency(my_graph)
```
