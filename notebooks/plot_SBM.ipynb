```python
import plotly.graph_objects as go


num_nodes = [50, 100, 500, 1000, 5000, 10000, 15000, 20000]
num_com = 10
num_models = 200


for i in range(len(num_nodes)):
    ndcgs = np.concatenate((list_all_n2v2r_non_binary[i],list_all_n2v2r_borda_non_binary[i],list_all_absDeDi_non_binary[i]), axis=None)
    method_indicator = (['n2v2r'] * ((num_com+1)*num_models) ) + (['n2v2r_borda'] * ((num_com+1)*num_models)) + (['absDeDi'] * ((num_com+1)*num_models))
    comms = [str(i) for i in range(1,num_com+1)]
    comms.insert(0,"total")

    comms_indicator = (comms * 3)*num_models

    df = pd.DataFrame(list(zip(ndcgs,  method_indicator, comms_indicator)),
                columns =['ndcgs', 'method', 'ordered communities'])
    
    df = df[((df['ordered communities']=='total'))]


    # fig = px.violin(df_restr, x="ordered communities", y="ndcgs", 
    #             color='method',  box=True, 
    #             height=400,  title= f'{num_nodes[i]} samples non binary',  color_discrete_sequence = ['#1b9e77','#d95f02','#7570b3'])



    # fig.show()

    fig = go.Figure()

    fig.add_trace(go.Violin(x=df['ordered communities'][ df['method'] == 'n2v2r_borda' ],
                            y=df['ndcgs'][ df['method'] == 'n2v2r_borda' ],
                            legendgroup='n2v2r_borda', scalegroup='n2v2r_borda', name='n2v2r_borda',
                            side='negative',
                            pointpos=-1.2,
                            line_color='#d95f02')
                )
    fig.add_trace(go.Violin(x=df['ordered communities'][ df['method'] == 'absDeDi' ],
                            y=df['ndcgs'][ df['method'] == 'absDeDi' ],
                            legendgroup='absDeDi', scalegroup='absDeDi', name='absDeDi',
                            side='positive',
                            pointpos=1.2,

                            line_color='#7570b3')
                )
                
    fig.update_traces(meanline_visible=True,spanmode = 'hard', box_visible=True,points='all', # show all points
                  jitter=0.05,)
    fig.update_layout(violingap=0, violinmode='overlay', title=f'{num_nodes[i]} nodes',
    xaxis_title="ordered communities",
    yaxis_title="ndcg",
    legend_title="Method",width=1000,
    height=600)

    fig.show()

```

```python
import plotly.graph_objects as go


num_nodes = [50, 100, 500, 1000, 5000, 10000, 15000, 20000]
num_com = 10
num_models = 200


for i in range(len(num_nodes)):
    dcgs = np.concatenate((list_all_n2v2r_binary[i],list_all_n2v2r_borda_binary[i],list_all_absDeDi_binary[i]), axis=None)
    method_indicator = (['n2v2r'] * ((num_com)*num_models) ) + (['n2v2r_borda'] * ((num_com)*num_models)) + (['absDeDi'] * ((num_com)*num_models))
    comms = [str(i) for i in range(1,num_com+1)]
    comms_indicator = (comms * 3)*num_models

    df = pd.DataFrame(list(zip(dcgs,  method_indicator, comms_indicator)),
                columns =['dcgs', 'method', 'ordered communities'])
    
    df = df[((df['ordered communities']=='1'))]


    # fig = px.violin(df_restr, x="ordered communities", y="ndcgs", 
    #             color='method',  box=True, 
    #             height=400,  title= f'{num_nodes[i]} samples non binary',  color_discrete_sequence = ['#1b9e77','#d95f02','#7570b3'])



    # fig.show()

    fig = go.Figure()

    fig.add_trace(go.Violin(x=df['ordered communities'][ df['method'] == 'n2v2r_borda' ],
                            y=df['dcgs'][ df['method'] == 'n2v2r_borda' ],
                            legendgroup='n2v2r_borda', scalegroup='n2v2r_borda', name='n2v2r_borda',
                            side='negative',
                            pointpos=-1.2,
                            line_color='#d95f02')
                )
    fig.add_trace(go.Violin(x=df['ordered communities'][ df['method'] == 'absDeDi' ],
                            y=df['dcgs'][ df['method'] == 'absDeDi' ],
                            legendgroup='absDeDi', scalegroup='absDeDi', name='absDeDi',
                            side='positive',
                            pointpos=1.2,
                            line_color='#7570b3')
                )
                
    fig.update_traces(meanline_visible=True,spanmode = 'hard', box_visible=True,points='all', # show all points
                  jitter=0.05,)
    fig.update_layout(violingap=0, violinmode='overlay', title=f'{num_nodes[i]} nodes',
    xaxis_title="ordered communities",
    yaxis_title="dcg",
    legend_title="Method",width=1000,
    height=600)

    fig.show()

```
